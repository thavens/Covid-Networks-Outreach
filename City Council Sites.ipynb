```python
from googlesearch import search
import time
import sqlite3
import re
from urllib.parse import urlsplit
```

```python
state = input('What state are these cities from? ')
with open("Cities.txt") as handle:
    cities = handle.readlines()
queries = []
for city in cities:
    city = city.strip()
    queries.append(city + ' city ' + state)
cities = zip(cities, queries)
cities
```

```python
conn = sqlite3.connect(r"D:/Programming/CovidNet/CovidNet.db")
sql_create_websites = """ CREATE TABLE IF NOT EXISTS websites (
                                        id INTEGER PRIMARY KEY AUTOINCREMENT UNIQUE,
                                        city TEXT,
                                        url TEXT,
                                        health TINYINT,
                                        searched TINYINT,
                                        state INTEGER
                                    ); """
c = conn.cursor()
c.execute(sql_create_websites)
```

```python
with open('Most Popular Sites.txt') as handle:
    pop_sites = handle.readlines()
def check(url):
    for i in pop_sites:
        if i.strip() in url:
            return False
    return True
```

```python
count = 0
long = time.time()
for city, query in cities:
    short = time.time()
    city = city.strip()
    health = 0
    com = True
    first = True
    url = ''
    for i in search(query, stop=5, pause=1.6):
        i = i.lower()
        if check(i):
            if city.replace(' ', '').lower() in i or city.replace(' ', '-').lower() in i:
                if check(url):
                    health = 1
                    url = i
                    break
            elif ".gov" in i:
                if check(url):
                    health = 1
                    url = i
                    break
    parts = urlsplit(url)
    base_url = f'{parts.scheme}://{parts.netloc}'
    if len(base_url) < 8:
        base_url = url
    c.execute(f'INSERT INTO websites (city, url, health, searched, state) VALUES (\"{city}\", \"{base_url}\", {health}, 0, {3})')
    print(city, base_url, health, round(time.time()-short, 2))
    if time.time() - 30 == long:
        conn.commit()
        print('commited to db')
conn.commit()
```
